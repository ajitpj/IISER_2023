# Some definitions
1. **Python** - Python uses an 'interpreter', which reads one line at a time, parses/interpretes the 'command', produces some result or output, and then moves on to the next line. 
2. **A command** - a native Python function/program that typically executes when invoked. It may require some input from the user and return some output. The output may simply be the completion of the process.
3. **A function** - A function is a user-defined command. When invoked, it will execute a set of commands and other functions. It may require an input and produce an output.
4. **Variables** - Data stored in computer memory; it can be recalled using the variable name.
5. **Objects** - Python treats every variable, data, function, etc. as an 'object'. This object can contain data as well as functions tied to the object and its data. These functions are also called 'methods'. Objects can be created by first defining a "class" that prescribes the properties and methods of the object.
6. **data structures** - As the name suggests, these are data containers specified by Python or one of the Python packages. These containers are designed to store specifc types of data.
7. **A script** - a sequence of commands that achieves a task that cannot be completed with just one command.
8. **A Python package/module** - This is a collection of functions written to work together for a specific purpose (e.g. image processing, machine learning, etc.). Packages are usually written by coding/scientific communities. Packages make Python more powerful and easy to use for specialized purposes.

In this class, we will mainly study how to use variables, functions, Python-created objects, and data structures to write useful scripts. Once you get comfortable with these tools, writing a nifty module or even a package should not sound daunting to you!

## 4. Indexing data structures
This is a critical concept. When you store many pieces of data in the same data structure, you need a way to retrieve specific data, excluding the data that you don't need. This data retrieval can be achieved by indexing. We will learn this in the context of Python lists, but the same principle will come into play when accessing data from pandas data frames and image arrays.

When you store many data in a list, Python assigns an index to each data. 

*   Indexing starts with 0 and ends with the length of your list minus 1.
*   Indexing is circular. Because the first item in the list has index 0, the last item will be -1 (imagine the last element of the list looping back to the first element to create a circle. Because the first element is indexed to 0, the last element will be indexed to -1, etc.)
*   To specify an index, you **must use square brackets**
*   To obtain items stored over a range of indices, use the syntax [first:last], where first and last are the indices of the first and last items within your range.
*   Use [first:last:stepsize] to retrieve items at discontinuous index values.
*   [:] retrieves all items in the list.
*   [::-1] retrieves all items in the reverse order (-1 becomes the stepsize)





```python
# Let's create a list with 10 random numbers
# TO create the list, we are going to use a Python module called 'random'.
import random 
# To use functions in this module, use the follwing syntax:
# random.function_name(function_arguments)
random.seed(1000) # For reproducing the same sequence of random numbers every time
# A convenient function to draw 10 random numbers between 0 and 100.
rand_list = random.sample(range(0, 100), 10) 
print(rand_list)
print("Length of the list: ", len(rand_list))
## To print the 7th number in this list:
# print("The 7th element in the list: ", rand_list[6]) 
## Must use square brackets for indexing in Python.
## To get data stored from the 2nd through 5th positions
print("Data from 2-5 are: ", rand_list[1:5])
## to replace the 7th element
rand_list[6] = 10000

# Indexing in Python is circular.
print(rand_list[:-1])
print(rand_list[::])

# Another way to assign the data stored in one list to another list
a = rand_list[:]
print(a)
```

#7. Conditional statements
What if you want to execute some commands only if certain condition or conditions are true (or false)? **Conditional** statements are a necessity in algorithms. 

The syntax for implementing code conditionally is very simple.

**What type of analysis could use conditional statements?**

```python
random_number = random.randint(0,10)
if random_number < 5:
  print("The random number generated, %s, is smaller than 5" %(random_number))
else:
  print("The random number generated, %s, is greater than or equal to 5" %(random_number))

# If you wanted to test multiple conditions, use the 'elif' clause.

if random_number < 5:
  print("The random number generated, %s, is smaller than 5" %(random_number))
  
elif random_number == 5:
  print("The random number generated, %s, is smaller than 5" %(random_number))

else:
  print("I don't care anymore")
```
